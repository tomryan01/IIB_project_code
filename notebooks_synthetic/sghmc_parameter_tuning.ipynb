```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../sghmc_outputs/synthetic/batch_size/sghmc_tuning_batch_size_hist.csv')
sizes = data['batch_size']
metrics = data['metric']
```

```python
fig, ax = plt.subplots()
sns.set_style('darkgrid')
ax = sns.lineplot(y=metrics, x=sizes)
ax.set_xlabel("batch size")
```

```python
import pandas as pd
import seaborn as sns
```

```python
bs_df = pd.read_csv('../sghmc_outputs/nn/batch_size/sghmc_tuning_batch_size.csv')
is_df = pd.read_csv('../sghmc_outputs/nn/int_samples/sghmc_tuning_int_samples.csv')
```

```python
batch_sizes = bs_df["batch_size"].to_numpy()
bs_metrics = bs_df["metric"].to_numpy()
int_samples = is_df["int_samples"].to_numpy()
is_metrics = is_df["metric"].to_numpy()
```

```python
bs_data = {"batch size" : batch_sizes, "histogram metric" : bs_metrics}
is_data = {"intermediate samples" : int_samples, "histogram metric" : is_metrics}
```

```python
sns.set_style('darkgrid')
sns.lineplot(bs_data, x='batch size', y='histogram metric', marker='o')
```

```python
sns.lineplot(is_data, x='intermediate samples', y='histogram metric', marker='o')
```
